```python
# gct (get currect taxis) = regresa el eje actual de dibujo 
# gcf (get currect figure)= regresa la figura del grafico

#definimos la funcion
def f(x):
    return np.exp(-x)*np.cos(2*np.pi*x)
```

# Texto en los graficos

```python
mu =100 # Media como 100
sigma = 20 # Desviacion tipica como 20 

x=mu+sigma*np.random.randn(10000)
```

```python
#Donde n=Numero de elementos / bins= Divisiones que hace de los datos, paches=Conjunto de datos respectivamente
# metodo hist, (<conjunto de datos>,<50 divisiones>,<normed= 1 Frecuencias relativas en eje vertical(donde 1=100 .025=.25) si se omite , las lineas son con base absoluta (barras= cuenta # de apariciones)>
#<facecolor= Color de barras>,<alpha=transparencia>)
n,bins,paches=  plt.hist(x,50, normed=1, facecolor="g", alpha=0.6)   

plt.xlabel("Cociente Intelectual", fontsize=14,color="green")  #Pintamos el texto del eje X
plt.ylabel("Probabilidad")          #Pintamos el texto del eje Y 
plt.title(r"Histograma de CI $N(\mu,\sigma)$")       #Pintamos el Titulo en el grafico

#Pintamos texto en la coordenada X=120,Y=0.015, con un registro markdown= r'<$ latex $>'
plt.text(120,0.015,r'$\mu=100,\ \sigma=20$')           

#Modificamos la numeracion del los ejes Ordenadas como Abcisas
#axis(<x_min>,<x_max>,<y_min>,<y_max>) pemrite cambiar la distribucion(numeracion) de los ejes X/Y
plt.axis([20,200,0,0.0250])  

#pintamos un grid en la grafica
plt.grid(True)

plt.show()
```

```python
#Pintar un objeto de tipo figura con parametros definidos, 
#figsize=tamao de figura de 10inch de ancho X 6inch de alto
#dpi= Calidad de pixeles por pulgada, 90px por inche
plt.figure(figsize=(8,6),dpi=90)
#solo pinta un 1 grafico en una caja de 1 Fila X 1 Columna
plt.subplot(1,1,1) 

#Definimos una serie de datos en el eje X
# x= np.arange(0,2*pi,0.01) SOLO MUESTRA 1 CURVA
x= np.arange(0,10*np.pi,0.01) #Muestra 5 curvas
y= np.cos(x)  #Calculamos el coseno de X (por cada valor dado en la matriz)

#plintamos el grafico, lw= Tamao de la linea del grafico
plt.plot(x,y,lw=2.0)

#pintamos un texto en una posicion especifica dentro del grafico, con la funcion annotate()
# annotate(<texto>,xy=<coordenada para una flecha x-y>,xytext=<coordenada para el texto x-y>,
# arrowprops<propiedades de las flechas con un diccionario de datos, donde facecolor=color, shrink=largo de flecha>)
plt.annotate("Maximo local", xy=(4*np.pi,1),xytext=(15,1.5),arrowprops=dict(facecolor="black",shrink=0.05))

#ajustamos el eje de las ordenadas con una separacion de -2 a 2
plt.ylim(-2,2)  
plt.show()
```

```python
#Podemoa anular las divisiones de los ejes en las graficas de tipo logisticas - plt.yscale("logit")
from matplotlib.ticker import NullFormatter
```

```python
mu= 0.5  # Media de 0.5
sd= 0.3  # Desviacion tipica de 0.3
y=mu+sd*np.random.randn(1000) #Un vector de numeros aleatorios
y=y[(y>0) & (y<1)]  # filtramos los datos del vector
y.sort()  #Ordnemas los datos del vector para que esten en orden ascendente
x=np.arange(len(y))
```

```python
#plt.figure(1)        #Una figura que contendra a los graficos 
#Ajustamos el tamaos del objeto figura que contiene los graficos en una ventana de 10 inch x 8 inch, para que no se subrepongan
plt.figure(figsize=(10,8)) 

plt.subplot(2,2,1)   #Una matriz de graficos de 2 filas, 2columnas, 1er grafico
plt.plot(x,y)
plt.yscale("linear")   #Escala lineal    
plt.xscale("linear")   #Escala lineal    
plt.title("Escala lineal")   #Grafica de tipo Escalar lineal  
plt.grid(True)  #lineas divisorias horizontale sy verticales

plt.subplot(2,2,2)
plt.plot(x,y)
plt.yscale("log")   #Escala log en ene eje de las Y  
#plt.xscale("log")  #Escala log en ene eje de las X   
plt.title("Escala Logaritmica")   #Escala logaritimica representada 
plt.grid()

plt.subplot(2,2,3)
plt.plot(x ,y -y.mean()) #Para representar el desplasamiento vertical log lo que se hace es sacar la madia para que este en 0 y no en 0.5
plt.yscale("symlog", linthreshy=0.01)  #Escala simetrica log en eje de las Y(linthreshy=divisiones logaritmicas a cada 0.01) 
plt.title("Escala logaritmica simetrica")
#logaritmo con crecimiento entre + y -, par que cobre sentido se debe incorporar valores -, para que este centrado a la media
plt.grid(True)

plt.subplot(2,2,4)
plt.plot(x,y)
plt.yscale("logit")  #Tipico de las curvas logisticas
plt.title("Escala logistica") 
plt.gca().yaxis.set_minor_formatter(NullFormatter())#Para eliminar los formatos en eje Y (evitamos que de vea ensimado )
plt.grid(True)

#Para ajustar de posicion ligeramente los graficos, 
#hspace= ajusta la altura para que los graficos no se sobrepongan entre ellos
#wspace= ajusta el ancho para que los graficos no se sobrepongan entre ellos
plt.subplots_adjust(top= 0.92, bottom=0.08, left=0.1, right=0.95, hspace=0.35, wspace=0.35) 
plt.show()

#HAY UN PROBLEMA CON EL FORMATO DE LA GRAFICA LOGIT, PARA EVITAR QUE SE VEA AMONTONADO HACEMOS IN IMPORTE
```

# Cambios en los ejes de dibujos

```python
# creamos un conjunto de datos con equidistribuidos con valores de -pi y +pi, con ayuda de la funcion 
# linspace(<inicio>,<fin>,<numero de datos>,<paa que empiece y termine en pi>)

x= np.linspace(-np.pi, np.pi,256,endpoint=True)  #Datos equidistribuidos
S, C= np.sin(x), np.cos(x)  #Asignamos los 256 valores equidistribuidos de x tanto en "S" como en "C"
```

```python
plt.plot(x, S, color="blue", linewidth=1.2, linestyle="-" )  # linewidth= Grosor de la linea
plt.plot(x, C, color="green", linewidth=1.2, linestyle="-" ) # linestyl= Estilo de la linea, pueden ser - , . , + , -. 

#Gestionamos las divisiones de los limites en los ejes de otra forma, los expandimos 10% mas con base a "X" y "S"
#plt.xlim(-4,4)  #xlim= Expandir los limites en en eje de las x
#plt.ylim(-1.5,1.5)  #ylim= Expandir los limites en el eje de las y de -1.5 a +1.5
plt.ylim(S.min()*1.1, S.max()*1.1)  #S.min()*1. Permite expandir un 10% mas el eje /S.max()*1.1 Permite expandir un 10%
plt.xlim(x.min()*1.1, x.max()*1.1)

#Podemos decidir donde aparecen los ticks, las marcas de las devisiones de los ejes y cuantas marcas o ticks
plt.xticks(np.linspace(-4,4,9, endpoint=True)) # 4=inicio, 4= fin ,9= separaciones
plt.yticks(np.linspace(-1,1,10, endpoint=True))
plt.show()
```

```python
plt.plot(x, S, color="blue",linewidth=1.2,linestyle="-" )  # linewidth= Grosor de la linea
plt.plot(x, C, color="green",linewidth=1.2, linestyle="-" ) # linestyl= Estilo de la linea, pueden ser - , . , + , -. 

#Gestionamos las divisiones de los limites en los ejes de otra forma, los expandimos 10% mas con base a "X" y "S"
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xlim(x.min()*1.1, x.max()*1.1)

#Podemos decidir donde aparecen los ticks, las marcas de las devisiones de los ejes y cuantos ticks
#plt.xticks(np.linspace(-4,4,9, endpoint=True)) # inica en -4, termina en 4, 8 lineas de numeracion
#plt.yticks(np.linspace(-1,1,10, endpoint=True)) # inica en -1, termina en 1, 10 lineas de numeracion

#Podemos generar un vector de ticks, para una distribucion donde cada division este marcada con valores como multiplos de PI
#y no con numeros enteros, ademas en el vector podemos asignar etiquetas en Latex como valores aplicados a los ticks
plt.xticks([-np.pi, -np.pi/2, 0,np.pi/2,np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$' ]) #Con latex

plt.yticks(np.linspace(-1,1,3, endpoint=True),
           ['-1','0','+1'])    #Sin latex


plt.show()
```

```python
plt.figure(figsize=(8,6))
plt.plot(x, S, color="blue",linewidth=1.2,linestyle="-" ,label="Seno" )  # linewidth= Grosor de la linea
plt.plot(x, C, color="green",linewidth=1.2, linestyle="-",label="Coseno") # linestyl= Estilo de la linea, pueden ser - , . , + , -. 

#Gestionamos las divisiones de los limites en los ejes de otra forma, los expandimos 10% mas con base a "X" y "S"
plt.ylim(S.min()*1.1, S.max()*1.1)
plt.xlim(x.min()*1.1, x.max()*1.1)

#Podemos decidir donde aparecen los ticks, las marcas de las devisiones de los ejes y cuantos ticks
#plt.xticks(np.linspace(-4,4,9, endpoint=True)) # inica en -4, termina en 4, 8 lineas de numeracion
#plt.yticks(np.linspace(-1,1,10, endpoint=True)) # inica en -1, termina en 1, 10 lineas de numeracion

#Podemos generar un vector de ticks, para una distribucion donde cada division este marcada con valores como multiplos de PI
#y no con numeros enteros, ademas en el vector podemos asignar etiquetas en Latex como valores de los ticks
plt.xticks([-np.pi, -np.pi/2, 0,np.pi/2,np.pi],
           [r'$-\pi$', r'$-\pi/2$', r'$0$', r'$+\pi/2$', r'+$\pi$' ])

plt.yticks(np.linspace(-1,1,3, endpoint=True),
           ['-1',' ','+1']) 

# Como modificar los ejes del dibujo actual, para esto se hace uso del metodo GCA()
ax=plt.gca()  #Con esta instruccion nos da acceso a la instancia actual del grafico 
ax.spines["right"].set_color("none")  #Da acceso a los bordes del grafico y remuebe el de la derecha
ax.spines["top"].set_color("none")  #Remueve el vorde de arriba

ax.xaxis.set_ticks_position("bottom")  #EL eje de las X lo colocamos abajo(esta habilitado por default), se necesita de forma explicita para acceder al eje
ax.spines["bottom"].set_position(('data',0))
ax.yaxis.set_ticks_position("left")
ax.spines["left"].set_position(("data",0))

plt.legend(loc="upper left")  #Instruccion legend= para colocar los label de lado Superior Izq 
#Definismos el punto a graficar 
x0= 2*np.pi/3

#Pintamo una linea recta en la grafica =  plot([<x1>,<x2>],[<y1>,<y2>])  -- Definimos punto de inicio y punto fin
plt.plot([x0,x0],[0,np.sin(x0)],color="blue", linewidth=2.0, linestyle="--")

#Dibujamos un punto en la grafica = scatter([<x1>,<x2>],[<y1>,<y2>], <Tamaos del punto>, <color del punto>)
plt.scatter([x0, ], [np.sin(x0), ], 50, color="blue")

#Escribimos un texto en la grafica y le damos formato.
#annotate(<TEXTO EN FORMATO LATEX>,<yx=(x,y) COORDENADA A COLOCAR TEXTO>, <xycoords="data" COORDENADA COPINCIDA CON LOS DATOS>)
#        <xytext=(10,10  DONDE SE UBICARA ES TEXTO FUERA DEL GRAFICO)> 
plt.annotate(r"$\sin(\frac{2\pi}{3})= \frac{\sqrt{3}}{2}  $"  , 
              xy=(x0, np.sin(x0)),  xycoords="data" ,
              xytext=(+40,+20),textcoords= "offset points",
              fontsize=16, arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))
#arrowprops= pemrite dibujar una figura en un punto con un,dict(<arrowstyle= figura>,connectionstyle=radianes - angulo)

## PUNTO EN  COORDENADA "Y"

plt.plot([x0,x0],[0,np.cos(x0)],color="green", linewidth=2.0, linestyle="--")
plt.scatter([x0, ], [np.cos(x0), ], 50, color="green")
plt.annotate(r"$\cos(\frac{2\pi}{3})= -\frac{\sqrt{1}}{2}  $"  , 
              xy=(x0, np.cos(x0)),  xycoords="data" ,
              xytext=(-90,-60),textcoords= "offset points",
              fontsize=16, arrowprops=dict(arrowstyle="->",connectionstyle="arc3,rad=.2"))
plt.show()
```
