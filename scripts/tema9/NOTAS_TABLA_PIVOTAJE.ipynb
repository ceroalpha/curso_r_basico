# Tabla Pivotaje

```python
import numpy as np
import pandas as pd
import seaborn as sb
```

```python
titanic= sb.load_dataset("titanic")  #importamoe el dataframe de datos
titanic.head() #mostamos los primeros 5 registros del ataframe
```

```python
#Agrupamos por la variable "SEX" y obtenemos la media de la columna "survied"
#Se muestra que 3 de cada 4 muejeres sobrevivio
titanic.groupby('sex')[['survived']].mean() 
```

```python
#Agrupamos por la variable "SEX" y "PCLASSS", y aplicamos un calculo de la media sobre los datos
#los datos aparecen apilados porque se seleccionaron w categorias 
titanic.groupby(['sex', 'pclass'])['survived'].aggregate('mean').unstack()
```

```python
#Agrupamos por la variable "SEX" y "PCLASSS", y aplicamos un calculo de la media sobre los datos
# unstack():  Permite que los datos aparescan en una tabla con formato de doble entrada 
#porcentaje de sobrevivencia por clase social, como el genero y la clase social afecto el indice de supervivencia
titanic.groupby(['sex', 'pclass'])['survived'].aggregate('mean').unstack()
```

```python
#Calculamos el % de personas sobreviviente
# pivot_table(<columna_evaluar>,<index=variabel en la fila index>, <columns=por columnas este esta variable>
titanic.pivot_table('survived', index='sex', columns='pclass')
```

```python
#podemos decidir transformar una variable en factor
#pd.cut(<dataframe>[age], bins=[lista de divisiones],include_lowest=<True /ajusta el intervalo para abrir mas los rangos>,
# precision=<1 /parametro de precision>, right=<False /Cierra los intervalos cerrado por la Izq y abierto por la Der.>)
age = pd.cut(titanic['age'], bins=[0,18,25, 35, 50, 65, 100], include_lowest=True, precision=1, right=False)
age
```

```python
#medir como sobrevivio la gente en funcion de la vaiable "edad",
#index["sex", age] : como filas sexo y edad y columns='class'  como columna la clase social 
titanic.pivot_table('survived', index=['sex', age], columns='class')
```

```python
#qcut = permite cortar la informacion por cuantiles, sin tener que indicar de forma explicita donde estan 
# pd.qcu(<dataframe>, q= <4 /cuantos cuantiles>, label=[lista de encabezados de los cuantiles])

fare = pd.qcut(titanic['fare'],q=4, labels= ["Q1", "Q2","Q3","Q4"])
fare # Calculamos la tarifa de los boletos
```

```python
#medir como En funcion de "survived" sobrevivio la gente, en funcion de filas vaiable "age"y "sex",
#pivot_table(<parametro_evaluar> index["sex",age]: como filas sexo y edad, columns=[fare, 'pclass'] como columna clase social y tarifa 

titanic.pivot_table('survived', index=['sex', age], columns=[fare, 'pclass'])
```

```python
#SINTAXIS DEL PIVOT
#DataFrame.pivot_table(data, values=None, index=None, columns=None, aggfunc="mean", fill_value=None, margins=False, dropna=True, margins_name = "All")
#data= conjunto de datos
#values=None, Los valores que tomara por defecto son None (valor de agrupacion)
#index=None, Filas que actuaran como index
#columns=None,  Columnas 
#aggfunc="mean", funcio de agre NA
#margins=False, Hace las marginales
#dropna=True, Si hay un dato NA no lo tomara para que aparesca en el interiore de dataframe
#margins_name = "All", nombres que aprqaceran si se calculan las marginales
```

```python
# En la funcion de "aggfunc" se le puede agregar un diccionario de funciones para que cada variable ap√±ique diferente medida
# values, no se suministro la columna por la cual se van a calcular los valores..  se suministro en el "aggfunc"
titanic.pivot_table(index = "sex", columns="class", 
                   aggfunc = {"survived":sum, "fare":"mean"})
```

```python
#Ejemplo con frecuencias marginales
titanic.pivot_table('survived', index="sex", columns="class", margins=True, margins_name="Total")
```
