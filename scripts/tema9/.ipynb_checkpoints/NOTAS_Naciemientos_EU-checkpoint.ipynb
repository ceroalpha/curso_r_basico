```python
'''
Creamos una Tabla de pivote con la siguiente extructura. 
    Numero total de necimiento por año y segmentada por genero 
pivot_table(<dataframe>,index=<variable indice>,columns=<columnas para el agregado> , aggfrunc=<Funcion aplicada a datos>)
'''
births.pivot_table('births', index= 'year', columns ='gender', aggfunc='sum')
```

```python
# Para el propio dataset agrupamos los datos por decadas, sin ocupar la funcion cut()

#Donde  de dataset "births" genero una variable con nombre 'var_decada' , donde divido cada  y me quedo con el cosiente 
# year//10  y lo multiplico para generar la decada   
births['var_decada']=  (births['year']//10) * 10 
```

```python
births.head()
```

```python
#Ahora apodemos generar la misma tabla pivite pero ahora por decada y no por año 

births.pivot_table('births', index= 'var_decada', columns ='gender', aggfunc='sum')
```

```python
'''
Cramos una tabla pivote con:
births.pivot_table(<como dataset>,index=<fila indice>,columns=<como columna>,aggfunc=<funcion de agregacion>  )
.plot(), permite generar la grafica de los datos 
'''

births.pivot_table('births',index='year', columns='gender', aggfunc='sum').plot()
plt.ylabel("Total de nacimientos por año") #Texto en el eje de las ordenadas
plt.xlabel("Intervalos de Años")  #Texto en el eje de las abcisas 
plt.show()   #pemrite eliminar la direccion de salida del grafico  "Text(0.5, 0, 'Intervalos de Años')"
```

```python
#Calculosmos los cuartiles Q1,Q2,Q3
#percentile(<dataset['variable']>,[<array de el numero de percentil>])
quartiles= np.percentile(births['births'],[25,50,75])
quartiles #Mostramos los cuartiles del numero de Nacimiento mensualde la tabla 
```

```python
#Estiamos la mediana que e slo mismo que el Cuartil 2 o 50% de los datos
media= quartiles[1]
IQR= quartiles[1]-quartiles[0] #Rango intercuartilico de los datos
sig= 0.75 *IQR  #Calculamos el sigma, la mitad del vigote en el digrama de caja y vigote (Desviacion tipica)
```

```python
#Analizando la siguiente instruccion, nos quedarnos con los datos que esten en un rango de mes +/- Rango intercuartilico
#Que al final son los datos menos dispersos y mas cercanos a la media, los que estan fuera se consideran atipicos
#la funcion query("<consulta>"),permite filtrar los datos en un dataset 

births= births.query("(births > @media  -5*@sig) & (births < @media +5*@sig )" ) 
```

```python
births.shape #Mostramos filas y columnas del dataset (filtrados como resultado de la instruccion anterior)
```

```python
births['day']= births['day'].astype(int) #Hacemos un casting de la variable numerica
```

```python
births.head() 
```

```python
#Creamos un objeto datetime, para el manejo de fechas en un formato que entienda python
#Creamos un dato que une YYYY+MM+DD y lo formateamos para poder trabajar con el dato tipo fecha y sus metodos
births.index= pd.to_datetime(10000*births.year+100*births.month+births.day, format="%Y%m%d")
```

```python
births.head() 
```

```python
#Creamos un  columna que contenga los dia de la semana, con la finalidad de graficar los nacimientos por dia
births['var_dayofWeek']=births.index.dayofweek
```

```python
births.head()
```

```python
#Agrupamos por dias de la semana 

births.pivot_table('births',index='dayofWeek',columns="var_decada",aggfunc="mean")    
```

```python
births.pivot_table('births',index='dayofWeek',columns="var_decada",aggfunc="mean").plot()
plt.gca().set_xticklabels([" ","L","M","M","J","V","S","D"])
plt.ylabel("Media de nacimiento por dia")
plt.show()

#Se logra identificar que los dias que menos son sabado y domingo 
```

```python
#Visualizamos el nacimiento por dias del año 
#De la columns 'births', obtengo las columnas mes  y dia, calculamos la media
births_by_date = births.pivot_table('births',index= [births.index.month,births.index.day],aggfunc="mean")
```

```python
#Tenemos la media de nacimiento por dia del mes N

births_by_date.head()
```

```python
#Para poder graficar los datos y que en los meses de berero el dia 29 no este vacio se crea una fecha temporal
#Donde, para cada  mes dia de la variable births_by_date.index, definida como indice, se crea un objeto con fecha 2020
births_by_date.index = [pd.datetime(2020, month, day) 
                        for (month, day) in births_by_date.index]
```

```python
births_by_date.head()
```

```python
#De la siguiente grafica podemos analizar que hay un par de dias del año la media de nacimientos baja
# 1 ENE, 4 JUL Indepenmdencia , 27-29 Accion de G. , 25 DIC Navidad, 31 DIC Fin de año
#Donde se deternina que los nacimientosson mas bajos en dias festivos

plt.figure(figsize=(12,4))
births_by_date.plot()
plt.show()
```
