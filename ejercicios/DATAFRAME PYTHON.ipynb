# Como crear un data frame desde Python

```python
#Creamos un dataframe con 10 filas y 3 columnas con valores aleatorios , su uso es como un df de prueba
df= pd.DataFrame({
    "A": np.random.rand(10),  
    "B": np.random.rand(10),
    "C": np.random.rand(10) 
})
```

```python
#Imprimimos el contenido del data frame
df
```

```python
# mean() = Calculamos la media por columna
df.mean()
```

```python
#mean(<axis=""columns>) = Permite calcular la medios pero ahora de cada una de las filas
df.mean(axis="columns")
```

```python
#Creamos un dicionario y lo asociamos a un data frame, pero ademas podemos colocarles nombres a las columnas con columns[]
df2= pd.DataFrame({
    "key": ["A","B","C","A","B","C"],  
    "values" :[1,2,3,4,5,6]
},  columns=["key","values"])
df2
```

```python
# Agregados para operaciones combinadas a traves de una TUPLA
# Donde, dataframe.groupby()  permite agrupar a una columna con tipo de dato tupla  y operar sobre esta 
df2.groupby("key")
```

```python
#aplicamos la funcion mean() a la columna KEY, esto permite colapsar la columna y calcular el mean() por cada valor de la tupla
df2.groupby("key").mean()
```

# Manipulacion de un datafreame

```python
df= pd.DataFrame({
    "key": ["A","B","C","A","B","C"],
    "value": range(6),
    "value2": np.random.randint(0,100,6)
})
```

```python
df
```

# aggregate({ ... , ... }) 

```python
#aggregate=  Permite agregar un vector de operaciones sobre el data frame, este vector de aplica por cada una de las variables
df.groupby('key').aggregate({'min',np.median, np.mean, max})
```

```python
#La funcion aggregate puede aplicar una operacion especifica por cada variable, por medio de un vector de operaciones
df.groupby('key').aggregate({
    'value' : 'min',
    'value2': 'max'
})
```

# filter( ... )

```python
# La funcion filter(), permite filtrar los datos contenidos en un un data frame a traves de una condicion logica
#1. Aplicamos la funcion std() al dataframe
df.groupby('key').std()
```

```python
#2. Creamos una funcion de filtrado que nos permita conservar los datos que en su STD sean mayores a 5 puntos
def funcion_filtrado(data):
    return data['value2'].std()>5
```

```python
#3. Hacemos uso de la funcion sobre el dataframe, implementando la funcion de filtrado
# como resultado solo debe mostrar los datos que cumplen la condicion
df.groupby("key").filter(funcion_filtrado)
```

# Transform(...)

```python
#La funcion se emplea para obtener una version reducida del datafranme aplicando operaciones o funciones, sin modificar el dato
df.groupby("key").transform(lambda data: data - data.mean())
```

# Apply

```python
# Nos permite aplicar cualquier funcion albitraria sobre el dataframe, el parametro necesita ser un DF y regresa un obj panda
#1. Definimosla funcion que se aplicara al dataframe

def norm_by_col2(data):
    data["value"] /= data["value2"].sum()
    return data
```

```python
#Hacemos uso de la funcion definida por medio de un apply
#como resultado se aplico una funcion sobre la columna 1, en funcion de agrupado por la llave
df.groupby("key").apply(norm_by_col2)
```

# Divicion de datos por columnas

```python
# Permite hacer una operacion con base a una clasificacion de los datos por medio de una lista 
# 1. Definimos la lista de dato
L=[0,1,0,1,0,2]
print(df.groupby("key").sum())
df
```

```python
#2. Aplicamos la suma sobre el dataframe, tomando las posicines de la lista y sumando de "L"
# los 0 , los 1 y los 2, acorde a la posicion en el data frame
df.groupby(L).sum()
```

```python
#Otra forma de optener el mismo resultado, CONSEPTUALMENTE ES LO MISMO, PERO EL PARAMETREO ES UNA VARIUABLE DEL DATAFRAME
df.groupby(df["key"]).sum()
```

```python
#df.set_index()=Permite eliminar el indice (numeracion de las filas) y tomar la variable "key" como la nueva nuemracion del df
df2= df.set_index("key")
df2
```

```python
#ahora que cambio el Indice podemos hacer un mapeo para trabajar con ese nuevo indice.
#Podemos trasnformar los agregados del indice con un diccionario y ajustarlo a operaciones de agrupado
mapping= {
    "A": "vovel",
    "B": "consonant",
    "C": "consonant"
}
df2.groupby(mapping).sum()
```

```python
# Tambien podemos pasar una funcion al indice. En este ejemplo la funcion lower
# Aplicamos el calculo del promedio y ahora el indice esta en minuscula
df2.groupby(str.lower).mean()
```

```python
#groupby(..)= Permite trabajar con mas de un indice y ademas aplicarles funciones
#Aplicamos la funcion minuscula al indice y al mapping 
df2.groupby([str.lower,mapping]).mean()
```

# Graficos desde dataframe y exportacion

```python
#Crear un data frame de forma manual
#Creamos el registro "aos" en un rango de valores "aos" entre 1900 a 2020
years=[year for year in range(1900, 2020)]
#Creamos un registro para cada uno de los registros un datos uniforme entre 0-100 con las datos de deads
deads=[ (y+np.random.uniform(0,100)-1850) for y in years]

df_datos=pd.DataFrame({
    "year": years,
    "deads": deads
})

df_datos.head()
```

```python
#hacemos uso de la funcion plot desde el dataframe para graficar el df
df_datos.plot(x="year",y="deads",figsize=(8,6))
```

```python
df_datos.plot(kind= "scatter", x="year",y="deads",figsize=(8,6))
```

```python
#Exportamos el grafico en formato png
#1. Guardmaos el data frame en un objeto
mi_plot= df_datos.plot(x="year",y="deads", figsize=(8,6))
```

```python
#2. Creamos un objeto de tipo figura, que son los que continen a un grafico en plot
mi_figura=mi_plot.get_figure()
#3. Exportamos el objeto tipo figura en formato png 
mi_figura.savefig("C:/Users/anton/OneDrive/Documents/Cursos/Estadistica descriptiva R y Python/Repositorio/Tareas/archivos/mi_figura_Sec9.png")
```
